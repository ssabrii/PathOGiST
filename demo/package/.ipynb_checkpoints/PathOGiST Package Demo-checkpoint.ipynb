`PathOGiST` has various submodules for clustering, creating distance matrices, file handling, and visualization.

```python
import pathogist
import pathogist.io
import pathogist.cluster
import pathogist.visualize
import pathogist.distance
```

Distance matrices can be opened like so:

```python
kwip_dist = pathogist.io.open_distance_file('pseudoTB13.dist')
mlst_dist = pathogist.io.open_distance_file('yersinia_mlst_dist.tsv')
snv_dist = pathogist.io.open_distance_file('yersinia_snv_dist.tsv')
```

We can easily visualize the distances in a histogram using the following function:

```python
pathogist.visualize.visualize(kwip_dist,'KWIP')
```

```python
pathogist.visualize.visualize(mlst_dist,'MLST')
```

```python
pathogist.visualize.visualize(snv_dist,'SNV')
```

It looks like some good thresholds for the KWIP, MLST and SNV data are 0.4, 400 and 2500, respectively. 
Let's perform some correlation clustering!

(My laptop only has the promotional version of CPLEX, so we'll first take a subset of the samples so that I can actually run the clustering algorithm.)

```python
sample_subset = snv_dist.index.values[0:10]
sample_subset
```

```python
kwip_dist.loc[sample_subset,sample_subset]
```

```python
snv_dist.loc[sample_subset,sample_subset]
```

```python
mlst_dist.loc[sample_subset,sample_subset]
```

```python
kwip_clustering = pathogist.cluster.correlation(kwip_dist.loc[sample_subset,sample_subset],threshold=0.4,all_constraints=False)
```

```python
mlst_clustering = pathogist.cluster.correlation(mlst_dist.loc[sample_subset,sample_subset],threshold=400.,all_constraints=False)
```

```python
snv_clustering = pathogist.cluster.correlation(snv_dist.loc[sample_subset,sample_subset],threshold=2500.,all_constraints=False)
```

```python
kwip_clustering.sort_index()
```

```python
snv_clustering.sort_index()
```

```python
mlst_clustering.sort_index()
```

Now that we have three clusterings, we can find their consensus clustering!

```python
distances = {'SNV':snv_dist,'MLST':mlst_dist,'KWIP':kwip_dist}
clusterings = {'SNV':snv_clustering,'MLST':mlst_clustering,'KWIP':kwip_clustering}
fine_clusterings = ['SNV']
consensus_clustering = pathogist.cluster.consensus(distances,clusterings,fine_clusterings,all_constraints=False)
```

```python
clusterings_summary = pathogist.cluster.summarize_clusterings(consensus_clustering,clusterings)
clusterings_summary.sort_index()
```

We can compare how much two clusterings differ from each other by calculating the adjusted rand index.

For example,

```python
pathogist.cluster.adjusted_rand_index(consensus_clustering,snv_clustering)
```

The consensus clustering and the SNV clustering are exactly the same (with different labels), which is to be expected since SNVs are the finest data type.
